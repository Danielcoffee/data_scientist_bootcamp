```python
import seaborn as sns
import pandas as pd
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
# can use marker, hue, scatter_kws,...
sns.lmplot(x='total_bill', y='tip', data=tips, markers=['o', 'v'], hue='sex', scatter_kws={'s': 50})
```
